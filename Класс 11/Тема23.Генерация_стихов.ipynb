# Анализ стихотворений
В этой же библиотеке предоставлен интерфейс для более детальной работы со стихами.

Подключаем Engine - основной модуль для работы, который создает объект для работы со стихами.

Загружаем в него методом `load`:
- модель ударений (stress),
- словарь (dict), который будет использоваться для стихосложения. 



```
# основной модуль
from rupo.api import Engine
```

```
# путь к модели ударений
stress_model_path='./rupo/rupo/data/stress_models/ru_main'
# путь к словарю
zalyzniak_dict_path='./rupo/rupo/data/dict'
# создаем объект-поэт
engine = Engine(language="ru")
# загружаем модели
engine.load(stress_model_path, zalyzniak_dict_path)
```

Генератор загружен и используется по умолчанию из директории RUPO.

Этот генератор сделан на основе LSTM сети:

```
engine.generator.model.model
```

Метод `get_stresses` сообщает, на какой символ приходится ударение (индексация с 0). Проверьте. Это важно знать, ведь стихи следуют определенному шаблону ударений.



```
print(engine.get_stresses("корова"))
print(engine.get_stresses("карабанга"))
```

Метод `get_word_syllables` разлагает слово по слогам.

```
# разлагаем на слога
print(engine.get_word_syllables("корова"))
print(engine.get_word_syllables("карабанга"))
```

Метод `is_rhyme` проверяет, образуют ли слова рифму.

```
print(engine.is_rhyme("корова", "здорова"))
print(engine.is_rhyme("ботинок", "полуботинок")) # Незнайка был не прав... 
```

Методом `classify_metre` можно узнать [стихотворный размер](https://ru.wikipedia.org/wiki/Стихотворный_размер):


Это *ямб* - размер, в котором ударения приходятся главным образом на четные слоги:

```
text = "Горит восток зарёю новой.\nУж на равнине, по холмам\nГрохочут пушки. Дым багровый\nКругами всходит к небесам."
engine.classify_metre(text)
```

А это *хорей* - ударения на нечетные слоги:

```
text ="Буря мглою небо кроет.\nВихри снежные крутя."
engine.classify_metre(text)
```

А это *дактиль* - один ударный и два подряд безударных слога за ним:

```
text="В рабстве спасенное\nСердце свободное —\nЗолото, золото\nСердце народное!"
engine.classify_metre(text)
```

Узнайте стихотворные размеры для других известных вам произведений.

```
text="Это просто текст\nбез всяких стихов\nчто тут\nможно определить?"
engine.classify_metre(text)
```

# Генерация стихотворений
Метод `generate_poem` генерирует стих по заданным критериям. Стихотворный размер задается в аргументе `metre_schema` строкой с символами `+` (ударный) и `-` (безударный).

Аргумент `rhyme_pattern` задает схему рифм по строчкам (т.е. какие именно строки должны рифмоваться) символами 'a', 'b'...

```
model_path='./rupo/rupo/data/generator_models' # путь к генератору
poem=engine.generate_poem(model_path=model_path, # путь к модели генератора
                     metre_schema='-+',# стихотворный размер (здесь - ямб)
                     rhyme_pattern='abab',# схема рифм (4 строки, рифмуются через одну)
                     n_syllables=8,# число слогов в строке
                     sampling_k=100,# число лучших кандидатов в следующее слово
                     temperature=2.0,# влияет на разнообразие 
                     seed=1000, # начало генератора случайных чисел
                     last_text='Конец уроку наступил')# последняя строка
print(poem)
```

Метод генерирует несколько подходящих слов - продолжений стиха, проверяет, что они подходят под условия (размер, количество слогов, рифма), выбирает лучшие. В жестких условиях может получиться так, что все слова будут отброшены, тогда возникнет сообщение об ошибке.

Если задан аргумент `last_text`, то метод будет пытаться закончить стих этой строкой (по умолчанию пустая ' ').

Аргумент `temperature` влияет на разнообразие слов-кандидатов. Чем больше температура, тем больше разнообразие, но и тем менее точно выполняются условия рифмы, размера и т.п. 

Чтобы при разных запусках метода изменялся ответ, изменяйте аргумент `seed`, который задает начало для генератора случайных чисел. Иначе результаты будут повторяться.
